```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
unemployment_men = pd.read_csv("Data/Men_Unemployment.csv")
unemployment_men=unemployment_men.set_index("DATE")
unemployment_men.head()

unemployment_women = pd.read_csv("Data/Women_Unemployment.csv")
unemployment_women=unemployment_women.set_index("DATE")
unemployment_women.head()
```

```python
# make empty chart and figure
figure,ax = plt.subplots()
#add data men to the line plot, ploted the chart
ax.plot(unemployment_men.index,unemployment_men["LNS14000001"])
#add data women to the line plot, ploted the chart
ax.plot(unemployment_women.index,unemployment_women["LNS14000002"])
#add title
ax.set_title("US Unemployment Rate Men and Women 1948-2024")
#label axis
ax.set_xlabel("Date")
ax.set_ylabel("Unemployment rate")
#cleanup date
ax.set_xticks(unemployment_men.index.values[0:-1:120])
plt.xticks(rotation = 90)
ax.legend(["Men", "Women"])

```
